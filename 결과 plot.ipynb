```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# result1 = open('CNN_results.txt', 'r', encoding='utf-8')
result = open('VGG19_results.txt', 'r', encoding='utf-8')
alex_str = result.read().strip()
alex_str = alex_str.split('\n')
for i in range(len(alex_str)):
    alex_str[i] = alex_str[i].split(', ')[3:]
    for j in range(2):
        alex_str[i][j] = alex_str[i][j].split(': ')
        alex_str[i][j] = float(alex_str[i][j][1])
alex_str = np.array(alex_str)
alex_str
```

```python
vgg_loss, vgg_acc = alex_str.T
print(vgg_loss)
print(vgg_acc)
```

```python
result = open('CNN_results.txt', 'r', encoding='utf-8')
cnn_str = result.read().strip()
cnn_str = cnn_str.split('\n')
for i in range(len(cnn_str)):
    cnn_str[i] = cnn_str[i].split(', ')[3:]
    for j in range(2):
        cnn_str[i][j] = cnn_str[i][j].split(': ')
        cnn_str[i][j] = float(cnn_str[i][j][1])
cnn_str = np.array(cnn_str)
cnn_str
```

```python
cnn_loss, cnn_acc = cnn_str.T
print(cnn_loss)
print(cnn_acc)
```

```python
fig, ax1 = plt.subplots()

X = np.array([i+1 for i in range(100)])
ax1.set_xlabel("Epochs")
ax1.set_ylabel('loss')
ax1.plot(X, cnn_loss, color='red')
ax1.plot(X, vgg_loss, color="darkorange")
line1 = ax1.plot(X, cnn_loss, color='red', label="CNN loss")
line2 = ax1.plot(X, vgg_loss, color="blue", label="VGG16 loss")
lines = line1+line2
labels = [i.get_label() for i in lines]
ax1.legend(lines, labels, loc="right")
plt.savefig("Fig2.png")
plt.show()
```

```python
fig, ax1 = plt.subplots()
ax1.set_xlabel("Epochs")
ax1.set_ylabel('accuracy')
ax1.plot(X, cnn_acc, color="blue")
ax1.plot(X, vgg_acc, color="deepskyblue")
line3 = ax1.plot(X, cnn_acc, color="red", label="CNN accuracy")
line4 = ax1.plot(X, vgg_acc, color="blue", label="VGG16 accuracy")

lines = line3+line4
labels = [i.get_label() for i in lines]
ax1.legend(lines, labels, loc="right")
# plt.savefig("Fig2.png")
plt.savefig("Fig3.png")
plt.show()
```
